```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('input/zomato.csv')
```

```python
df=df.dropna(axis=0,how="any")
```

# Plot resturant type vs votes(most preferable resturant type)

```python
plt.barh(df.rest_type.tail(10),df.votes.tail(10))
```

# Plot resturant type vs resturant name

```python
plt.barh(df.rest_type.tail(10),df.name.tail(10)).plot(figsize=(14,0.5))
```

# Plot cuisines vs resturant type

```python
plt.bar(df.rest_type.head(10),df.cuisines.head(10))
```

# Ploting cuisines mostly orderd from specific resturants

```python
plt.bar(df.name.head(10),df.cuisines.head(10))
```
