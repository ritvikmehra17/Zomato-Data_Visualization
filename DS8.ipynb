```python
import numpy as np 
import pandas as pd 

import plotly.plotly as py
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)
import plotly.graph_objs as go


import matplotlib.pyplot as plt

import os
```

```python
df=pd.read_csv('input/zomato.csv')
```

```python
df=df.dropna(axis=0,how="any")
```

```python
trace1 = go.Scatter(
                    x = df.votes.head(10),
                    y = df.name.head(10),
                    mode = "lines",
                    name = "res_name",
                    marker = dict(color = 'rgba(16, 112, 2, 0.8)'),
                    text= df.cuisines.head(10))
# Creating trace2

data = [trace1]
layout = dict(title = 'res_name and cuisine vs votes',
              xaxis= dict(title= 'votes',ticklen= 5,zeroline= False)
              
             )
fig = dict(data = data, layout = layout)
iplot(fig)
```

```python
df["votes"]/=100
v=df.votes.head(10).astype(float)
v
```

```python
df.rate = df.rate.str.replace('/5','')
```

```python
df.rate = df.rate.str.replace('NEW','0').replace('-','')
df.rate = df.rate.str.strip()
df.rate =df.rate.replace('','0')
df.rate=df.rate.astype(float)
```

```python


trace1 =go.Scatter(
                    x = v,
                    y = df.rate.head(10),
                    mode = "markers",
                    name = "res_name",
                    marker = dict(color = 'rgba(255, 128, 255, 0.8)'),
                    text= df.name.head(10))
# creating trace2
trace2 =go.Scatter(
                    x = v,
                    y = df.rate.head(10),
                    mode = "markers",
                    name = "cuisine",
                    marker = dict(color = 'rgba(255, 128, 2, 0.8)'),
                    text= df.cuisines.head(10))
# creating trace3
trace3 =go.Scatter(
                    x = v,
                    y = df.rate.head(10),
                    mode = "markers",
                    name = "res_type",
                    marker = dict(color = 'rgba(0, 255, 200, 0.8)'),
                    text= df.rest_type.head(10))
data = [trace1, trace2, trace3]
layout = dict(title = 'rate vs votes',
              xaxis= dict(title= 'votes',ticklen= 5,zeroline= False),
              yaxis= dict(title= 'rank',ticklen= 5,zeroline= False)
             )
fig = dict(data = data, layout = layout)
iplot(fig)
```

```python


# create trace1 
trace1 = go.Bar(
                x = df.location.head(10),
                y = df.phone.head(10),
                name = "rates",
                marker = dict(color = 'rgba(255, 174, 255, 0.5)',
                             line=dict(color='rgb(0,0,0)',width=1.5)),
                text = df.rate.head(10))
# create trace2 
trace2 = go.Bar(
                x = df.location.head(10),
                y = df.phone.head(10),
                name = "cuisine",
                marker = dict(color = 'rgba(255, 255, 128, 0.5)',
                              line=dict(color='rgb(0,0,0)',width=1.5)),
                text = df.cuisines.head(10))
data = [trace1, trace2]

layout = go.Layout(barmode = "group")
fig = go.Figure(data = data, layout = layout)
iplot(fig)
```

```python

num_students_size  = [float(each.replace(',', '.')) for each in df.index.head(10)]
international_color = [float(each) for each in df.international]
data = [
    {
        'y': df.rate.head(10),
        'x': df.votes.head(10),
        'mode': 'markers',
        'marker': {
            'color': international_color,
            'size': num_students_size,
            'showscale': True
        },
        "text" :  df.name.head(10)    
    }
]
iplot(data)
```

```python
trace1 = go.Scatter(
    x=df.name.head(10),
    y=df.votes.head(10),
    name = "votes"
)
trace2 = go.Scatter(
    x=df.name.head(10),
    y=df.rate.head(10),
    xaxis='x2',
    yaxis='y2',
    name = "citations"
)
trace3 = go.Scatter(
    x=df.name.head(10),
    y=df.cuisines.head(10),
    xaxis='x3',
    yaxis='y3',
    name = "income"
)
trace4 = go.Scatter(
    x=df.name.head(10),
    y=df.location.head(10),
    xaxis='x4',
    yaxis='y4',
    name = "total_score"
)
data = [trace1, trace2, trace3, trace4]
layout = go.Layout(
    xaxis=dict(
        domain=[0, 0.45]
    ),
    yaxis=dict(
        domain=[0, 0.45]
    ),
    xaxis2=dict(
        domain=[0.55, 1]
    ),
    xaxis3=dict(
        domain=[0, 0.45],
        anchor='y3'
    ),
    xaxis4=dict(
        domain=[0.55, 1],
        anchor='y4'
    ),
    yaxis2=dict(
        domain=[0, 0.45],
        anchor='x2'
    ),
    yaxis3=dict(
        domain=[0.55, 1]
    ),
    yaxis4=dict(
        domain=[0.55, 1],
        anchor='x4'
    ),
    title = 'Research, citation, income and total score VS World Rank of Universities'
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```
