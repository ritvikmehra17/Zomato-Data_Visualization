```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('input/zomato.csv')
```

```python
df=df.dropna(axis=0,how="any")
```

```python
df["location"].value_counts()
```

# Most frequently used resturant

```python
order_counts=df["name"].value_counts()
order_counts.head(20).plot(kind="barh")
plt.show()
```

```python
order_counts.head().plot(kind='bar') #top five resturants
```

# which area has most north indian food orders

```python
type_of_cuisine="North Indian"
cuisine_filter=df['cuisines']==type_of_cuisine
cuisine=df[cuisine_filter]
cuisine
```

```python
df['cuisines']=="North Indian"
```

```python
is_north_indian=df['cuisines']=="North Indian"
```

```python
cuisine=df[is_north_indian]
cuisine['location'].value_counts().dropna()
```

```python
cuisine_counts=cuisine['location'].value_counts().head()
total_cuisine_counts=df['location'].value_counts().head()
```

```python
cuisine_counts/total_cuisine_counts
```

```python
cuisine_counts/total_cuisine_counts.astype(float).plot(kind="bar")
```
